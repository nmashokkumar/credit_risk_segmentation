## Model Building

```python
df.columns
```

```python
cols =['LIMIT_BAL', 'SEX', 'EDUCATION', 'MARRIAGE', 'AGE', 'PAY_0', 'PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6', 'BILL_AMT1', 'BILL_AMT6',
             'PAY_AMT1','PAY_AMT6']

feature = df[cols]
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(feature)
```

## KMeans Elbow Method

```python
inertia = []
K = range(1, 11)

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)  # Use standardized data
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8,5))
plt.plot(K, inertia, 'bo-', markersize=8)
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia (WCSS)')
plt.title('Elbow Method For Optimal k')
plt.grid(True)
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(X_scaled)
```

```python
print(df['Cluster'].value_counts())
```

```python
cluster_profile = df.groupby('Cluster').mean()
print(cluster_profile[['LIMIT_BAL', 'AGE', 'PAY_0', 'PAY_2', 'PAY_AMT1', 'PAY_AMT6']])
```

```python
plt.figure(figsize=(8,6))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=df['Cluster'], cmap='viridis', s=10)
plt.title('KMeans Clusters (k=3)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
cluster_labels = {0: 'Low Risk', 1: 'Medium Risk', 2: 'High Risk'}
plt.figure(figsize=(8,6))
plt.scatter(X_pca[:,0], X_pca[:,1], c=df['Cluster'], cmap='viridis', s=10)
plt.title('KMeans Clusters Visualized with PCA (k=3)')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.show()
```
